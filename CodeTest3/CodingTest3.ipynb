```python
import pandas as pd
import numpy as np
import os
import yfinance as yf
import datetime as dt
from sklearn.model_selection import TimeSeriesSplit
import pandas_ta as ts
import skopt
```

```python
#MSTR invest in BTC after August, 2020, I want to see if this has some relation with bitcoin.
start_date = dt.datetime(2020, 9, 1)
end_date = dt.datetime.now()

mstr_data = yf.Ticker("MSTR").history(start=start_date, end=end_date, interval="1d")

mstr_data.index = mstr_data.index.date

print(mstr_data.head())
```

```python
def concatenate_csv_files(folder_path):
    file_list = [f for f in os.listdir(folder_path)]
    
    df_list = []
    
    for file_name in file_list:
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path, header = None)
        df_list.append(df)
    
    concatenated_df = pd.concat(df_list, ignore_index=True)
    return concatenated_df
```

```python
folder_path = os.path.expanduser('~/Desktop/Desktop')
bin_dataframe = concatenate_csv_files(folder_path)
bin_dataframe.dropna(inplace = True)

datetime_index = pd.to_datetime(bin_dataframe.iloc[:, 0].apply(int), unit='ms')
bin_dataframe.index = datetime_index
bin_dataframe = bin_dataframe.drop(0, axis = 1)
bin_dataframe.rename(columns = {1:'open_btc',2:'high_btc',3:'low_btc',4:'close_btc',5:'volume_btc',6:'close_time',
                                7:'quote_volume',8:'count',9:'taker_buy_volume',
                                10:'taker_buy_quote_volume',11:'ignore'}, inplace = True)
bin_dataframe.dropna(inplace=True)
print(bin_dataframe.head())
```

```python
df = bin_dataframe.join(mstr_data)
df.ffill(inplace=True) #fill the stock data on weekends 
```

```python
tscv = TimeSeriesSplit(n_splits=2)
train = []
test = []
for trainindex, testindex in tscv.split(df):
    train.append(trainindex)
    test.append(testindex)
```

```python
for trainidx in train:
    tr = df.iloc[trainidx]
    print(np.corrcoef([tr['close_btc'], tr['Close']]))
```

```python
#we see a very strong positive correlation between btc and mstr
def strategy(df,sma1, sma2, mom, tc = 0):
    rst = df.copy()
    rst['return'] = np.log(rst['close_btc'] / rst['close_btc'].shift())
    rst['mreturn'] = rst['return'].rolling(mom).mean()
    rst['sma1'] = np.log(rst['Close'].rolling(sma1).mean())
    rst['sma2'] = np.log(rst['Close'].rolling(sma2).mean())
    
    #when either stocks performs better recently or the momentum is positive, we set the position to long
    rst['signal'] = np.where((rst['sma1'] >= rst['sma2']) | (rst['mreturn']>0), 1, 0)
    #when both stocks performs worse recently or the momentum is negative, we set the position to short
    rst['signal'] = np.where((rst['sma1'] <= rst['sma2']) & (rst['mreturn']<0), -1, rst['signal'])
    
    rst['strategy'] = rst['signal'].shift() * rst['return']
    trades = rst['signal'].diff().fillna(0) != 0
    rst['strategy'][trades] -= tc
    rst.dropna(inplace = True)
    rst['creturn'] = rst['return'].cumsum().apply(np.exp)
    rst['cstrategy'] = rst['strategy'].cumsum().apply(np.exp)
    return rst
```

```python
rst1 = strategy(df.iloc[train[0]], 5, 20,10, 0.001)
rst1[['creturn', 'cstrategy']].plot()
#this randomly chosen strategy underperforms.
```

```python
#now optimize the return through the data corresponding to the last index in train.

SPACE = [skopt.space.Integer(5, 30, name = 'sma1', prior = 'uniform'),
        skopt.space.Integer(40,100, name = 'sma2', prior = 'uniform'),
        skopt.space.Integer(5,80, name = 'mom', prior = 'uniform')]

@skopt.utils.use_named_args(SPACE)
def objective(**params):
    rst1 = strategy(df.iloc[train[-1]], tc = 0.001, **params)
    a = rst1['cstrategy'].iloc[-1]
    rst2 = strategy(df.iloc[train[0]], tc = 0.001, **params)
    b = rst1['cstrategy'].iloc[-1]
    return (-a*2/3-b*1/3)/2 #we give larger weight to the last train data as it is more recent

results = skopt.forest_minimize(objective, SPACE,
                               base_estimator='RF', n_calls = 100,
                               n_initial_points=10, verbose = False, n_jobs=-1)
best_params = results.x
best_return = -results.fun  

print(f"Best parameters: {best_params}")
print(f"Best sharpe: {best_return}")
```

```python
def sharpe_calculate(df):
    return df.mean()/df.std()
```

```python
import matplotlib.pyplot as plt
rst2 = strategy(df.iloc[train[-1]], 21, 55, 56, 0.001)
rst2.plot(y = ['creturn', 'cstrategy'])
plt.axhline(y=1, color='red', linestyle='-')

sharpe = sharpe_calculate(rst2['strategy'].dropna())
print(f'Sharpe ratio: {sharpe}')

sharpe = sharpe_calculate(rst2['return'].dropna())
print(f'Sharpe ratio no strategy: {sharpe}')
```

```python
#according to the above, we choose sma1 = 21, sma2 = 55, mom = 56
#we test on the data corresponding to the last index of test
rst3 = strategy(df.iloc[test[-1]],21, 55, 56, tc = 0.001)
rst3[['creturn', 'cstrategy']].plot()
#we see that our strategy performs well when th emarket goes down, in the test part, the market goes up, this results in a lower performance.
```

```python
#we see from sharpe ratio that the startegy performs poorly when the market goes up.
sharpe = sharpe_calculate(rst3['strategy'].dropna())
print(f'Sharpe ratio: {sharpe}')

sharpe = sharpe_calculate(rst3['return'].dropna())
print(f'Sharpe ratio no strategy: {sharpe}')
#Compared with above, this strategy seems to perform better when the market underperforms.
#the reason for the underperformance is due to the different performance of the market for the training and test set.
#the two sharpe ratio is close.
```

```python
capital = 100000
rst3['capital'] = rst3['strategy'].dropna().cumsum().apply(np.exp)* capital
rst3['cummax'] = rst3['capital'].cummax()
rst3['drawdown'] = rst3['cummax'] - rst3['capital']
mdd = rst3['drawdown'].max()
print(f'Maximum Drawdown: {mdd}')


rst3['capital2'] = rst3['return'].dropna().cumsum().apply(np.exp)* capital
rst3['cummax2'] = rst3['capital2'].cummax()
rst3['drawdown2'] = rst3['cummax2'] - rst3['capital2']
mdd = rst3['drawdown2'].max()
print(f'Maximum Drawdown no strategy: {mdd}')
#the maximum drawdown for strategy is smaller.
#Given a close sharpe ratio and a much smaller maximum drawdown, 
#the strategy may not give the maximum profit, but it is stabler. 
#the startegy can constantly generate profit when the market is not good.
```
