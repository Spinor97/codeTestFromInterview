```python
import pandas as pd
import numpy as np
import requests
from binance.client import Client
import os
```

```python
def fetch_historical_data(market, start_date, end_date, count=200):
    url = "https://api.upbit.com/v1/candles/days"
    headers = {"Accept": "application/json"}
    data = []
    
    current_date = end_date
    
    while current_date >= start_date:
        params = {
            'market': market,
            'to': current_date,
            'count': count
        }
        response = requests.get(url, headers=headers, params=params)
        
        if response.status_code == 200:
            candles = response.json()
            data.extend(candles)
            # Update current_date to the date of the last candle received
            current_date = candles[-1]['candle_date_time_kst']
        else:
            break
    
    return pd.DataFrame(data)
```

```python
btc_kim = fetch_historical_data('KRW-BTC','2021-01-01T00:00:00Z', '2023-01-31T23:59:59Z')
btc_kim.rename(columns = {'prev_closing_price': 'close_kim'}, inplace = True)
btc_kim.index = [pd.to_datetime(x[:10]) for x in btc_kim.candle_date_time_utc]
btc_kim = btc_kim.drop(['candle_date_time_utc', 'candle_date_time_kst'], axis = 1)
btc_kim = btc_kim.sort_index()
```

```python
def concatenate_csv_files(folder_path):
    file_list = [f for f in os.listdir(folder_path)]
    
    df_list = []
    
    for file_name in file_list:
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path, header = None)
        df_list.append(df)
    
    concatenated_df = pd.concat(df_list, ignore_index=True)
    return concatenated_df
```

```python
folder_path = os.path.expanduser('~/Desktop/BTSUSDTD')
bin_dataframe = concatenate_csv_files(folder_path)

datetime_index = pd.to_datetime(bin_dataframe.iloc[:, 0], unit='ms')
bin_dataframe.index = datetime_index
bin_dataframe = bin_dataframe.drop(0, axis = 1)
bin_dataframe.rename(columns = {4 : 'Close_bin'}, inplace = True) #The fourth column after transforming the datetime index is close.
print(bin_dataframe.head())
```

```python
df = bin_dataframe.join(btc_kim)
df = df[['Close_bin', 'close_kim']]
#we calculate the return of upbit and we shift the return to the next day
df['return_kim'] = np.log(df['close_kim'] / df['close_kim'].shift()).shift() 
#we do not shift the return for binance
df['return_bin'] = np.log(df['Close_bin'] / df['Close_bin'].shift())
df.dropna(inplace = True)
print(df.head())
```

```python
df_train = df[:len(df)//2].copy()
df_test = df[len(df)//2:].copy()
```

```python
def return_calculate_xy(x,y, df):
    data = df.copy()
    data['signal_bin'] = np.where(data['return_kim']>=x, 1, 0)
    data['signal_bin'] = np.where(data['return_kim']<=-1*y, -1, data['signal_bin'])
    data['strategy'] = data['signal_bin'] * data['return_bin']
    return data

rst1 = return_calculate_xy(0.1,0.05,df_train.copy())
print(rst1.head())
print(rst1['strategy'].dropna().cumsum().apply(np.exp))
rst1[['return_bin', 'strategy']].dropna().cumsum().apply(np.exp).plot()
#The strategy does not perform well with (0.1, 0, 05)
```

```python
def sharpe_calculate(df):
    return df.mean()/df.std()
```

```python
M = df_train['return_bin'].max()
m = df_train['return_kim'].min()

x_values = np.linspace(0, M, 11)
y_values = np.linspace(0, abs(m), 21)
sharpe_df = pd.DataFrame(index=x_values, columns=y_values)
for x in x_values:
    for y in y_values:
        tr = return_calculate_xy(x,y, df_train.copy())
        sharpe = sharpe_calculate(tr['strategy'].dropna())
        sharpe_df.at[x,y] = sharpe
sharpe_df = sharpe_df.astype(float)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(25,25))
ax1 = sns.heatmap(sharpe_df, annot=True, cmap='coolwarm', fmt='.4f')
plt.title('Sharpe Ratio Heatmap')
plt.xlabel('Y  - Short Limit')
plt.ylabel('X  - Long Limit')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig("sharpe_ratio_heatmap.png")
plt.show()
```

```python
#we see from graph that the maximum of sharpe value is achieved near the right bottom.
rst2 = return_calculate_xy(0.7754,0.1287,df_train.copy())
print(rst2.head())
print(rst2['strategy'].dropna().cumsum().apply(np.exp))
rst2[['return_bin', 'strategy']].dropna().cumsum().apply(np.exp).plot()
```

```python
#we choose the parameter with largest return, given the sharpe ratio large
return_df = pd.DataFrame(index=x_values, columns=y_values)
for x in x_values:
    for y in y_values:
        tr = return_calculate_xy(x,y, df_train.copy())
        returns = tr['strategy'].dropna().cumsum().apply(np.exp).iloc[-1]
        return_df.at[x,y] = returns
return_df = return_df.astype(float)
```

```python
plt.figure(figsize=(26,25))
ax2 = sns.heatmap(return_df, annot=True, cmap='coolwarm', fmt='.4f')
plt.title('Return Heatmap')
plt.xlabel('Y  - Short Limit')
plt.ylabel('X  - Long Limit')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig("return_ratio_heatmap.png")
plt.show()
```

```python
#considering sharpe ratio, we may choose the parameter to be (0.7754, 0.1287) now we test this parameter with df_test
rst3 = return_calculate_xy(0.7754,0.1287,df_test.copy())
print(rst3.head())
print(rst3['strategy'].dropna().cumsum().apply(np.exp))
rst3[['return_bin', 'strategy']].dropna().cumsum().apply(np.exp).plot()
#we see from the graph that this strategy may not outperform, but it can help us avoiding large loss.
```

```python
#considering return, we choose parameters to be (0.09, 0)
rst4 = return_calculate_xy(0.09, 0,df_test.copy())
print(rst4.head())
print(rst4['strategy'].dropna().cumsum().apply(np.exp))
rst4[['return_bin', 'strategy']].dropna().cumsum().apply(np.exp).plot()
#we see from the graph that this parameters combination has higher return.
#this may suggest that we should not be extreme on one criteria.
```

```python
#calculate sharpe ratio and CAGR and maximum draw down for (0.0775, 0.05148)
sharpe = sharpe_calculate(rst4['strategy'].dropna())
print(f'Sharpe ratio: {sharpe}')

#we assume a year has 365 days
CAGR = rst4['strategy'].dropna().cumsum().apply(np.exp).iloc[-1] ** (1/((len(df_train)/365))) - 1
print(f'CAGR: {CAGR}')

#suppose the capital is 100000
capital = 100000
rst4['capital'] = rst4['strategy'].dropna().cumsum().apply(np.exp)* capital
rst4['cummax'] = rst4['capital'].cummax()
rst4['drawdown'] = rst4['cummax'] - rst4['capital']
mdd = rst4['drawdown'].max()
print(f'Maximum Drawdown: {mdd}')
```
