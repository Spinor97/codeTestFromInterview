```python
import pandas as pd
import numpy as np
import requests
from binance.client import Client
import os
```

```python
def concatenate_csv_files(folder_path):
    file_list = [f for f in os.listdir(folder_path)]
    
    df_list = []
    
    for file_name in file_list:
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path, header = None)
        
        #from 2022, the csv has a name for each column, we need to remove them for he convenience of later use. 
        if str(df.iloc[0, 1]).isalpha() == True:
            df.drop([0], inplace = True)
            
        df_list.append(df)
    
    concatenated_df = pd.concat(df_list, ignore_index=True)
    return concatenated_df
```

```python
folder_path = os.path.expanduser('~/Desktop/BTCUSDTHour')
bin_dataframe = concatenate_csv_files(folder_path)
bin_dataframe.dropna(inplace = True)

datetime_index = pd.to_datetime(bin_dataframe.iloc[:, 0].apply(int), unit='ms')
bin_dataframe.index = datetime_index
bin_dataframe = bin_dataframe.drop(0, axis = 1)
bin_dataframe.rename(columns = {1:'open',2:'high',3:'low',4:'close',5:'volume',6:'close_time',
                                7:'quote_volume',8:'count',9:'taker_buy_volume',
                                10:'taker_buy_quote_volume',11:'ignore'}, inplace = True)
bin_dataframe.dropna(inplace=True)
print(bin_dataframe.head())
```

```python
print(bin_dataframe.isnull().sum())
```

```python
def alpha_A(df):
    typical_price = (df['close'].apply(float) + df['high'].apply(float) + df['low'].apply(float)) / 3
    cumulative_volume = df['volume'].apply(float).rolling(24).sum()
    cumulative_typical_price = (typical_price * df['volume'].apply(float)).rolling(24).sum()
    vwap = cumulative_typical_price / cumulative_volume
    alpha_A = np.sqrt(df['high'].apply(float) * df['low'].apply(float)) - vwap
    return alpha_A

def alpha_B(df):
    alpha_B = -1 * ((df['low'].apply(float) - df['close'].apply(float))
                    * (df['open'].apply(float) ** 5)) / np.maximum(((df['low'].apply(float) - df['high'].apply(float)) 
                                                                    * (df['close'].apply(float)) ** 5),0.001)
    return alpha_B

def alpha_C(df):
    denominator = np.maximum(df['high'].apply(float) - df['low'].apply(float), 0.0001)
    alpha_C = (df['close'].apply(float) - df['open'].apply(float)) / denominator
    return alpha_C

bin_dataframe['alpha_A'] = alpha_A(bin_dataframe)
bin_dataframe['alpha_B'] = alpha_B(bin_dataframe)
bin_dataframe['alpha_C'] = alpha_C(bin_dataframe)
```

```python
bin_dataframe['return'] = np.log(bin_dataframe['close'] / bin_dataframe['close'].shift())
bin_dataframe.dropna(inplace=True)
bin_train = bin_dataframe[:len(bin_dataframe)//2]
bin_test = bin_dataframe[len(bin_dataframe)//2:]
```

```python
#we calculate the correlation between return and the three alphas
category = ['A', 'B', 'C']
correlations = {}
for idx in category:
    correlations[idx] = np.corrcoef([bin_train['close'], bin_train[f'alpha_{idx}']])

for i in category:
    print(correlations[i])
```

```python
#from above we see that the close is positively correlated with alphaB, so we try a strategy based on the sma performance of alphaB
#sma1 < sma2
def SMA(df,sma1, sma2):
    rst = df.copy()
    rst['sma1'] = rst['alpha_B'].rolling(sma1).mean()
    rst['sma2'] = rst['alpha_B'].rolling(sma2).mean()
    rst['signal'] = np.where(rst['sma1'] >= rst['sma2'], 1, -1)
    rst['strategy'] = rst['signal'].shift() * rst['return']
    rst.dropna(inplace = True)
    rst['creturn'] = rst['return'].cumsum().apply(np.exp)
    rst['cstrategy'] = rst['strategy'].cumsum().apply(np.exp)
    return rst
```

```python
# we choose sma1 = 3 and sma2 = 10 randomly
rst1 = SMA(bin_train, 7, 8)
rst1[['creturn', 'cstrategy']].plot()
#the return is poor, but it is natural as we choose randomly the parameter.
```

```python
def performance(sma):
    return SMA(bin_train, sma[0], sma[1])['cstrategy'].iloc[-1]

x_val = np.linspace(1, 24, 24).astype(int)
y_val = np.linspace(1, 24, 24).astype(int)
ret_bin = pd.DataFrame(index=x_val, columns=y_val)
for x in x_val:
    for y in y_val:
        if y > x:
            ret = performance((x, y))
            ret_bin.at[x,y] = ret
        else:
            ret_bin.at[x,y] = 0
ret_bin = ret_bin.astype(float)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(25,25))
ax1 = sns.heatmap(ret_bin, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Return Heatmap')
plt.xlabel('SMA2')
plt.ylabel('SMA1')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig("return_heatmap.png")
plt.show()
```

```python
#we notice that sma1 = 18 and sma2 = 20 we have highest return.
rst2 = SMA(bin_train, 18, 20)
rst2[['creturn', 'cstrategy']].plot()
#other than this pair, we have other pairs also performing well: (16,17), (18,19), (19, 20), (23, 24), (22, 24), (21, 24), (17, 24)
#we calculate their sharpe values and choose the pair with the largest value.
def sharpe_calculate(df):
    return df.mean()/df.std()

selected = [[18, 20], [16,17], [18,19], [19, 20], [23, 24], [22, 24], [21, 24], [17, 24]]
sharpes = {}
mdds = {}
for idx, pair in enumerate(selected):
    rst = SMA(bin_train, pair[0], pair[1])
    sharpes[idx] = sharpe_calculate(rst['strategy'].dropna())
    capital = 100000
    rst['capital'] = rst['strategy'].dropna().cumsum().apply(np.exp)* capital
    rst['cummax'] = rst['capital'].cummax()
    rst['drawdown'] = rst['cummax'] - rst['capital']
    mdd = rst['drawdown'].max()
    mdds[idx] = mdd
    
print(sharpes)
print(mdds)
# we notice that pair[0] and pair[4] have almost the same sharpe ratio, but pair[4] has smaller maximum draw down, so we choose pair[4]
```

```python
#now we test on the test data with (22, 24)
rst3 = SMA(bin_test, 22, 24)
rst3[['creturn', 'cstrategy']].plot()
plt.savefig("return_ratio_heatmap(22,24).png")
#compared with the train data and test data, we see that this startegy outperforms when the market goes down.
```

```python
sharpe = sharpe_calculate(rst3['strategy'].dropna())
print(f'Sharpe ratio: {sharpe}')

capital = 100000
rst3['capital'] = rst3['strategy'].dropna().cumsum().apply(np.exp)* capital
rst3['cummax'] = rst3['capital'].cummax()
rst3['drawdown'] = rst3['cummax'] - rst3['capital']
mdd = rst3['drawdown'].max()
print(f'Maximum Drawdown: {mdd}')

CAGR = rst3['strategy'].dropna().cumsum().apply(np.exp).iloc[-1] ** (1/((len(bin_test)/365))) - 1
print(f'CAGR: {CAGR}')
```
